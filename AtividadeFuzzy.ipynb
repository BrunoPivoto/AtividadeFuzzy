```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
velocidade = ctrl.Antecedent(np.arange(0, 181, 1), "velocidade")
massa = ctrl.Antecedent(np.arange(0, 2.5, 0.1), "massa")
pressao = ctrl.Consequent(np.arange(0, 1.1, 0.1), "pressao")

velocidade.automf(5)
nomes_velocidades = ["MB", "BA", "ME", "AL", "MA"]
velocidade.automf(number=5, names=nomes_velocidades)

massa.automf(5)
nomes_massas = ["MP", "PE", "ME", "GR", "MG"]
massa.automf(number=5, names=nomes_massas)

pressao.automf(3)
nomes_pressao = ["MI", "ME", "EL"]
pressao.automf(number=5, names=nomes_pressao)

velocidade.view()
massa.view()
pressao.view()
```

```python
rule1 = ctrl.Rule(velocidade["MB"] & massa["MP"], pressao["MI"])
rule2 = ctrl.Rule(velocidade["BA"] & massa["MP"], pressao["MI"])
rule3 = ctrl.Rule(velocidade["ME"] & massa["MP"], pressao["MI"])
rule4 = ctrl.Rule(velocidade["AL"] & massa["MP"], pressao["ME"])
rule5 = ctrl.Rule(velocidade["MA"] & massa["MP"], pressao["ME"])

rule6 = ctrl.Rule(velocidade["MB"] & massa["PE"], pressao["MI"])
rule7 = ctrl.Rule(velocidade["BA"] & massa["PE"], pressao["MI"])
rule8 = ctrl.Rule(velocidade["ME"] & massa["PE"], pressao["MI"])
rule9 = ctrl.Rule(velocidade["AL"] & massa["PE"], pressao["ME"])
rule10 = ctrl.Rule(velocidade["MA"] & massa["PE"], pressao["ME"])

rule11 = ctrl.Rule(velocidade["MB"] & massa["ME"], pressao["MI"])
rule12 = ctrl.Rule(velocidade["BA"] & massa["ME"], pressao["MI"])
rule13 = ctrl.Rule(velocidade["ME"] & massa["ME"], pressao["ME"])
rule14 = ctrl.Rule(velocidade["AL"] & massa["ME"], pressao["ME"])
rule15 = ctrl.Rule(velocidade["MA"] & massa["ME"], pressao["ME"])

rule16 = ctrl.Rule(velocidade["MB"] & massa["GR"], pressao["ME"])
rule17 = ctrl.Rule(velocidade["BA"] & massa["GR"], pressao["ME"])
rule18 = ctrl.Rule(velocidade["ME"] & massa["GR"], pressao["EL"])
rule19 = ctrl.Rule(velocidade["AL"] & massa["GR"], pressao["EL"])
rule20 = ctrl.Rule(velocidade["MA"] & massa["GR"], pressao["EL"])

rule21 = ctrl.Rule(velocidade["MB"] & massa["GR"], pressao["ME"])
rule22 = ctrl.Rule(velocidade["BA"] & massa["GR"], pressao["ME"])
rule23 = ctrl.Rule(velocidade["ME"] & massa["GR"], pressao["EL"])
rule24 = ctrl.Rule(velocidade["AL"] & massa["GR"], pressao["EL"])
rule25 = ctrl.Rule(velocidade["MA"] & massa["GR"], pressao["EL"])
```

```python
velocidade_ctrl = ctrl.ControlSystem([rule1, rule2, rule3, rule4, rule5, rule6,rule7, rule8, rule9,
                                      rule10, rule11, rule12, rule13, rule14, rule15,rule16, rule17,
                                      rule18, rule19, rule20, rule21, rule22, rule23,rule24, rule25])
velocidadeTip = ctrl.ControlSystemSimulation(velocidade_ctrl)
```

```python
velocidadeTip.input["velocidade"] = 155
velocidadeTip.input["massa"] = 1.5

velocidadeTip.compute()

resposta = velocidadeTip.output["pressao"]
print("A presso  de: ", round(resposta,4), "atm")
```
